#### Categorize the labels into Numaric or Attack

```python
df_copy["binary_label"]=df_copy["labels"].apply(lambda x: "Noraml" if x=="normal" else "Attack")
```

```python
df_copy["binary_label"]
```

```python
df_copy["binary_label"].value_counts()
```

### Categories of Attack

```python
df_copy["labels"].unique()
```

```python
attack_mapping={
    ## DOS attack
    'back':'DOS',"land":"DOS",'neptune':'DOS','pod':'DOS','smurf':'DOS',
    ## Probe attack
    'satan':'Probe','ipsweep':'Probe','nmap':'Probe','portsweep':'Probe',
    ## U2R
    'buffer_overflow':'U2R','loadmodule':'U2R','perl':'U2R','rootkit':'U2R',
    ##R2L
    'ftp_write':'R2L','guess_passwd':'R2L','imap':'R2L','warezclient':'R2L','multihop':'R2L','phf':'R2L','spy':'R2L',
    ### Normal
    'normal':'Normal'
    }

df_copy["attack_type"]=df["labels"].map(attack_mapping)
```

```python
df_copy["attack_type"]
```

```python
df_copy["attack_type"].value_counts()
```

```python
df_copy.head()
```

```python
import matplotlib.pyplot as plt
df_copy["binary_label"].value_counts().plot.pie(autopct="%1.2f%%")
```

- Above pie plot states that there is a half of network behaves in defferent way 
- There is a attacking from third person of 46%

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=df_copy["count"] ,y=df_copy["serror_rate"],hue=df_copy["attack_type"],alpha=0.8)
```

- source error rate with 1 struggling with DOS(Denial Of Service)
- source error rate with 0.2 struggling with Probe attack
- from the above plot we conclude that we transform the less count of bytes in the form of packets throgh the network

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df_copy,x='attack_type',order=df["attack_type"].value_counts().index,palette='Set2')
plt.title("Attack Type Distrubution")
plt.tight_layout()
plt.show()
```

- DOS Attack is high among all types of attacks (U2R,R2L,PROBE)
